```python
import pandas as pd
import numpy as np
from tmdbv3api import TMDb
import json
import requests
```

```python
tmdb = TMDb()
tmdb.api_key = '7f8389d09aba49383e25f1bda263445f'
```

```python
class MoviesExtraction:
    def __init__(self,link):
        self.link = link
    def read_movie2018(self):
        df1 = pd.read_html(self.link, header=0)[2]
        df2 = pd.read_html(self.link, header=0)[3]
        df3 = pd.read_html(self.link, header=0)[4]
        df4 = pd.read_html(self.link, header=0)[5]
        df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
        return df
    
    def read_movie2019(self):
        df1 = pd.read_html(self.link, header=0)[3]
        df2 = pd.read_html(self.link, header=0)[4]
        df3 = pd.read_html(self.link, header=0)[5]
        df4 = pd.read_html(self.link, header=0)[6]
        df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
        return df
    
    def read_movie2020(self):
        df1 = pd.read_html(self.link, header=0)[3]
        df2 = pd.read_html(self.link, header=0)[4]
        df3 = pd.read_html(self.link, header=0)[5]
        df4 = pd.read_html(self.link, header=0)[6]
        df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
        return df
    
    def read_movie2021(self):
        df1 = pd.read_html(self.link, header=0)[1]
        df2 = pd.read_html(self.link, header=0)[2]
        df3 = pd.read_html(self.link, header=0)[3]
        df4 = pd.read_html(self.link, header=0)[4]
        df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True, sort=False)
        return df
    
```

# 2018

```python
data18 = MoviesExtraction("https://en.wikipedia.org/wiki/List_of_American_films_of_2018")
df18 = data18.read_movie2018()
df18.head()
```

```python
df18.shape
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
```

```python
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
        
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
    
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])

def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
    
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df18['genres'] = df18['Title'].apply(get_genre)
```

```python
df18['director_name'] = df18['Cast and crew'].apply(get_director)
df18['actor_1_name'] = df18['Cast and crew'].apply(get_actor1)
df18['actor_2_name'] = df18['Cast and crew'].apply(get_actor2)
df18['actor_3_name'] = df18['Cast and crew'].apply(get_actor3)
```

```python
df18
```

```python
data_2018 = df18[['Title', 'genres', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name']]
data_2018
```

```python
data_2018['combination'] = data_2018['actor_1_name'] + ' ' + data_2018['actor_2_name'] + ' '+ data_2018['actor_3_name'] + ' '+ data_2018['director_name'] +' ' + data_2018['genres']
```

```python
data_2018
```
