```python
#importing libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('dataset/final_movie_dataset.csv')
data.head()
```

```python
data = data[['movie_title','director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres', 'combination']]
data.head()
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(data['combination']) 
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
movie_user_likes = 'Avatar'
movie_index = data.loc[data['movie_title']== movie_user_likes].index[0]

lst = list(enumerate(cosine_sim[movie_index]))
lst = sorted(lst, key = lambda x:x[1] ,reverse=True)
lst = lst[1:11] # excluding first item since it is the requested movie itself
l = []
for i in range(len(lst)):
    a = lst[i][0]
    l.append(data['movie_title'][a])
print(l)
```
